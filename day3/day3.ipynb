```python
import pandas as pd
import numpy as np
```

```python
with open("input.txt") as f:
    input = [line.strip() for line in f]
```

```python
# make dataframe

# input = ['191111911', '111199111', '1111188', '99111', '1111119']

combos = []
for i in range(len(input)):
    input[i] = list(map(int, list(input[i])))
    largest = np.argmax(input[i][:-1])
    
    tens = input[i][largest]
    unit = input[i][np.argmax(input[i][largest+1:]) + largest + 1]

    combos.append(int(str(tens) + str(unit)))

# combos
print(sum(combos))

# bankheaders = [f"bank_{i}" for i in range(len(input[0]))]
# df = pd.DataFrame(input, columns=bankheaders)
# df.sort_values(by=bankheaders, inplace=True, ascending=False)
# df.head()
```

```python
# input = ['987654321111111', '811111111111119', '234234234234278', '818181911112111']
combos = []



for item in input:
    switchedon = 1
    combo = str()

    item = list(map(int, list(item)))
    largest = np.argmax(item[:-12])
    combo += str(item[largest])
    remaining = item[largest+1:]
    
    while switchedon < 12:
        if switchedon + len(remaining) == 12:
            combo += ''.join(map(str, remaining))
            break
        max_idx = len(remaining) - (12 - switchedon) + 1
        largestremaining = np.argmax(remaining[:max_idx])
        combo += str(remaining[largestremaining])
        remaining = remaining[largestremaining+1:]
        switchedon += 1


    combos.append(int(combo))

# combos
print(sum(combos))
```
