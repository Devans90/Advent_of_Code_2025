```python
input = """67562556-67743658,62064792-62301480,4394592-4512674,3308-4582,69552998-69828126,9123-12332,1095-1358,23-48,294-400,3511416-3689352,1007333-1150296,2929221721-2929361280,309711-443410,2131524-2335082,81867-97148,9574291560-9574498524,648635477-648670391,1-18,5735-8423,58-72,538-812,698652479-698760276,727833-843820,15609927-15646018,1491-1766,53435-76187,196475-300384,852101-903928,73-97,1894-2622,58406664-58466933,6767640219-6767697605,523453-569572,7979723815-7979848548,149-216""".split(',')

rawdata = []

def split_string(s, n):
    return list(map(''.join, zip(*[iter(s)]*n)))

def checkforrepeat(itemlist):
    repeats = []
    length = len(itemlist) / 2
    if length % 1 == 0.0:
        fullset = split_string(itemlist, int(length))
        splitset = set(fullset)
        if len(splitset) < len(fullset):
            repeats.append(itemlist)
            return itemlist
        

# def checkforrepeat(itemlist):
#     repeats = []
#     length = 2
#     while len(itemlist) // length > 1:
#         for _ in range(len(itemlist) // length):
#             splitset = set(split_string(itemlist, length))
#             fullset = split_string(itemlist, length)
#             if len(splitset) < len(fullset):
#                 repeats.append(itemlist)
#                 return itemlist
#         length += 1
#     return None

for item in input:
    itemsplit = item.split('-')
    full = list(range(int(itemsplit[0]), int(itemsplit[1])))
    rawdata.append(full)

repeatingvalues = []

for fullitem in rawdata:
    for item in fullitem:
        item = checkforrepeat(str(item))
        if item is not None:
            repeatingvalues.append(int(item))

repeatingvalues
```
